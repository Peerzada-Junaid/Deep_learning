```python
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt 
```

```python
img= Image.open("../datasets/ice.jpg")
```

```python
img.size
```

```python
small=img.resize(size=(400,300))
```

```python
small
```

```python
img= np.array(small).mean(axis=2)
```

```python
plt.imshow(img,cmap="gray")
plt.show()
```

```python
padded_image= np.pad(img, pad_width=1,mode="constant")
```

```python
padded_image.shape
```

```python
plt.imshow(padded_image, cmap="gray")
```

```python
filt= np.array([[1,1,1],[0,0,0],[-1,-1,-1]])    # vertical filter 
```

```python
result=np.zeros_like(img)      # we are creating an array of zeros similar to the size of the img
```

```python
for row in range (result.shape[0]):
    for col in range (result.shape[1]):
        cut = padded_image[row:row+3,col:col+3]
        result[row, col]=(cut*filt).mean()
```

```python
plt.imshow(result, cmap="gray")
```
