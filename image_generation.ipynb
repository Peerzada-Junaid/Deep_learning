```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img_dog=load_img("../datasets/german-shepherd.jpg",target_size=(300,400))
img_cat=load_img("../datasets/Persian_Cat.jpg", target_size=(300, 400))
```

```python
type(img_dog)
```

```python
array_dog = np.array(img_dog)
array_cat = np.array(img_cat)
```

```python
array_dog.shape,array_cat.shape
```

```python
batch = np.array([array_dog,array_cat])
```

```python
batch.shape
```

```python
generator = ImageDataGenerator(rotation_range=10,zoom_range=0.1,horizontal_flip=True,shear_range=1.1)
```

```python
gen=generator.flow(batch, batch_size=10)
```

```python
out=gen.next()
```

```python
for gen_image in out:
    plt.figure()
    plt.imshow(gen_image/255)   # or as type (int)
```
